## Perguntas - Obesity among adults by country, 1975-2016

```python
df_obesity = pd.read_csv("obesity_cleaned.csv")
```

- Limpe os dados do DataFrame, criando uma coluna de nome 'Obesity' que conterá os valores de obesidade. Transforme em float as colunas que porventura foram importadas como string.

```python
df_obesity.info()
```

```python
df_obesity['Obesity (%)'].describe()
```

```python
#exclui "no data do indice de obesidade"
df = df_obesity[df_obesity['Obesity (%)'] != "No data"]
df
```

```python
# Divide a coluna usando separador
df[['nivel','x','x1']] = df['Obesity (%)'].str.split('[', expand=True)
```

```python
df.head(3)
```

```python
df['nivel'] = df['nivel'].astype(float)
```

```python
df['nivel'].describe()
```

```python
#seleciona um subset do dataset

df = df[['Country','Year','Sex','nivel']]
df
```

- Qual o percentual médio de obesidade por sexo no mundo no ano de 2015?

```python
gr_mean = df.groupby(['Sex'])['nivel'].mean()
gr_mean
```

- Quais são os 5 países com a maior e a menor taxa de aumento nos índices de obesidade no período observado?

```python
# Agrupa País para cada ano e calcula a média do índice
gr_world = (df.groupby(['Country', 'Year'])['nivel'].mean().reset_index().sort_values(by='Year', ascending=True))
gr_world['growth_rate'] = (gr_world.groupby(['Country'])['nivel'].pct_change()*100)

gr1 = gr_world.groupby('Country')['growth_rate'].mean().sort_values(ascending=False)


```

```python

print("Países com as maiores taxas de crescimento:")
print(gr1.head(5))
```

```python

print("Países com as menores taxas de crescimento:")
print(gr1.tail(5))
```

- Quais os países com maiores e menores níveis percetuais de obesidade em 2015?

```python
df2015 = df[df['Year'] == 2015]
gr_2015 = df2015.groupby(['Country'])['nivel'].mean().reset_index().sort_values(by='nivel', ascending=False)
```

```python
print("Países com maiores níveis de obesidade em 2015:")
print(gr_2015.head(10))
```

```python
print("Países com menores níveis de obesidade em 2015:")
print(gr_2015.tail(10))
```

- Qual a diferença média percentual de obesidade entre sexos ao longo dos anos para o Brasil?

```python
df_brasil =  df[df['Country'] == 'Brazil']
gr_year_brasil = df_brasil.groupby(['Year','Sex'])['nivel'].mean().reset_index()

#transforma tabela pivot
gr_year_brasil_pivot = gr_year_brasil.pivot(index ='Year',columns='Sex', values='nivel')

#diferença % homens e mulheres
gr_year_brasil_pivot['diff%'] = ((gr_year_brasil_pivot['Female'] - gr_year_brasil_pivot['Male'])/gr_year_brasil_pivot['Male'])*100
```

```python
gr_year_brasil_pivot.head(7)
```

```python
#diferença média percentual
gr_year_brasil_pivot['diff%'].mean()
print(f'A diferença média percentual é de {gr_year_brasil_pivot["diff%"].mean():.2f}%')
```

- Você conseguiria plotar um gráfico mostrando a evolução da obesidade para ambos sexos no mundo?

```python

df_both = df.groupby(['Year','Sex'])['nivel'].mean().reset_index()

#transforma tabela pivot
df_both_pivot = df_both.pivot(index ='Year',columns='Sex', values='nivel')

#diferença % homens e mulheres
df_both_pivot['diff%'] = ((df_both_pivot['Female'] - df_both_pivot['Male'])/df_both_pivot['Male'])*100
```

```python
fig = px.line(df_both_pivot,
              x=df_both_pivot.index,
              y=['Male','Female'],
              height=400,
              width=1000,
              title= "Evolução da Obesidade no Mundo")
fig.show()


```

## Perguntas - GDP Per Person (1901-2011)

```python
df_gdp = pd.read_csv("gdp.csv", decimal=".")
```

- Limpe o conjunto de dados, convertendo strings em datas ou float, quando necessário.

```python
df_gdp.info()
```

```python
#transforma variaveis
df_gdp['Year1'] = pd.to_datetime(df_gdp['Year'])

#Poda o nome da coluna
df_gdp['GDP_pp'] = df_gdp[' GDP_pp '].str.strip()

#Retira caracteres indesef=jados da coluna
df_gdp['GDP_pp'] = df_gdp['GDP_pp'].str.replace(',', '').astype(float)

```

- Você conseguiria informar o primeiro valor registrado de cada país?

```python
#Primeiro valor de cada país
first_value = df_gdp.groupby('Country')['GDP_pp'].first().reset_index()
first_value
```

- Informe as regiões com maiores crescimentos de PIB per capita no século passado.

```python
sec_20  = df_gdp[df_gdp['Year1'] < '2001-01-01']
sec_20.head(5)
```

```python
gr_sec_20 = sec_20.groupby(['Region','Year1'])['GDP_pp'].mean().reset_index().sort_values(by='Year1', ascending=True)
gr_sec_20.head(5)
```

```python
gr_sec_20['growth_rate'] = (gr_sec_20.groupby(['Region'])['GDP_pp'].pct_change()*100)
gr_sec_20.head(5)
```

```python
Taxa_crescimento = gr_sec_20.pivot(index='Region', columns='Year1', values='growth_rate')
Taxa_crescimento
```

```python
valor_gdp = gr_sec_20.pivot(index='Region', columns='Year1', values='GDP_pp')
valor_gdp
```

```python


import plotly.express as px

gr_sec_20.sort_values(by='Region', ascending=False)
gr_sec_20['Region'] = gr_sec_20['Region'].replace('Middle East, North Africa, and Greater Arabia', 'MENA')


fig = px.bar(gr_sec_20,
             x="Region",
             y='growth_rate',
             hover_name='Region',
             hover_data={'growth_rate': True, 'Region': False, 'Year1': False},
             color='growth_rate',
             color_continuous_scale=px.colors.sequential.Blues,
             barmode='group',
             orientation='v',
             category_orders = {'Region': gr_sec_20['Region']},
             range_y=[0, 70],
             width=1500,
             height= 500,
             title='Regiões com maiores crescimentos de PIB per capita no século 20',
             animation_frame = 'Year1')
fig.show()

```

- Preecha os anos ausentes em cada país com uma estimativa, baseada na diferença entre o próximo registro e o anterior.

```python
#Frequancia de paises
df_gdp.head(2)
df_country = df_gdp.groupby(['Country'])['Year1'].count().reset_index()
print("Todos os países aparecem na cronologia com a mesma frequência. Não há períodos faltantes ")
print(df_country)
```

- Você conseguiria criar um mapa do gdp ou da obesidade no mundo ao longo dos anos?

```python
#Agrupando
gr_world = df_gdp.groupby(['Region', 'Country', 'Year1'])['GDP_pp'].mean().reset_index()

# Importando a biblioteca plotly express
import plotly.express as px



fig = px.scatter_geo(gr_world,
                     locations='Country',
                     locationmode='country names',
                     color='GDP_pp',
                     hover_name='Country',
                     hover_data={'GDP_pp': True, 'Country': False, 'Year1': False},
                     size='GDP_pp',
                     size_max=50,
                     projection='natural earth',  # Projeção natural da Terra
                     animation_frame='Year1',  # Animação por ano
                     title='Mapa do GDP no Mundo')

fig.show()
```

- Há uma relação entre níveis de obesidade e gdp per capita?

```python
df_pib = df_gdp.groupby(['Country'])['GDP_pp'].mean().reset_index()

df_obesity = df.groupby(['Country'])['nivel'].mean().reset_index()
```

```python
df_gdp_ob = pd.merge(df_pib, df_obesity, on='Country')
df_gdp_ob.head(5)
```

```python

fig = px.scatter(df_gdp_ob,
                 x='GDP_pp',
                 y='nivel',
                 hover_name='Country',
                 hover_data={'GDP_pp': True, 'nivel': True},
                 color='nivel',
                 color_continuous_scale=px.colors.sequential.Viridis,
                 title='Relação entre níveis de obesidade e gdp per capita')
fig.show()
```

```python
fig = px.density_heatmap(
      data_frame=df_gdp_ob,
      x='nivel',
      y='GDP_pp',
      color_continuous_scale='dense',
       width=800,
      height=400,
      title='Densidade da relação entre níveis de obesidade e gdp per capita'

)
fig.show()

#Matriz de correlação
df_matrix = df_gdp_ob[['GDP_pp','nivel']]
corr_matrix = df_matrix.corr()
print('Correlação entre variáveis:')
print(corr_matrix)
```
