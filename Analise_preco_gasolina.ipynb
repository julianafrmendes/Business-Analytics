## Neste projeto, trabalharemos com dois arquivos csvs separados, contendo informações sobre o preço da gasolina no Brasil.

### 1. Carregue os conjuntos de dados "gasolina_2000+.csv" e "gasolina_2010+.csv" em dois DataFrames diferentes e combine-os em um único DataFrame.

```python
import pandas as pd

df1 = pd.read_csv('gasolina_2000+.csv')
df2 = pd.read_csv('gasolina_2010+.csv')
df = pd.concat([df1, df2])
```

### 4. Você deve ter percebido que as colunas DATA INICIAL e DATA FINAL estão formatadas como string. Utilizando o método pd.to_datetime(), converta ambas colunas para Timestamp / Datetime.

```python
df['DATA INICIAL'] = pd.to_datetime(df['DATA INICIAL'])
df['DATA FINAL'] = pd.to_datetime(df['DATA FINAL'])
df.info()
```

### 5. Crie uma nova coluna para representar o ano e o mês(aaaa-mm), utilizando a coluna DATA FINAL como referência.

DESAFIO: Crie uma tabela contendo uma serie temporal com a diferença absoluta e percentual entre os valores mais baratos e caros. Apresente também ao lado os estados na qual os maiores e menores preços foram registrados.

```python
#descriçao de valores
df['PREÇO MÉDIO REVENDA'].describe()
```

```python
gasolina_comum
```

```python
min = gasolina_comum.groupby(['DATA INICIAL'])['PREÇO MÉDIO REVENDA'].max().reset_index()
max = gasolina_comum.groupby(['DATA INICIAL'])['PREÇO MÉDIO REVENDA'].min().reset_index()

amp= max.merge(min, on='DATA INICIAL')

```

```python
#renomear colunas
amp.rename(columns={'PREÇO MÉDIO REVENDA_x': 'min', 'PREÇO MÉDIO REVENDA_y': 'max'}, inplace=True)
```

```python
#clacula diferença percentual
amp['diff%'] = (amp['max'] - amp['min']) / amp['min'] * 100
```

```python
#calcula diferença absoluta
amp['diff_abs'] = abs(amp['max'] - amp['min'])
```

```python
#Separa os valores de fiferença absoluta minímos
max_diff = amp.sort_values(by='diff_abs', ascending = False).head(15)
max_diff
```

```python
#agrupamento  para buscar o estado de valores mínimos
gr5 =  gasolina_comum.groupby(['DATA INICIAL', 'ESTADO'])['PREÇO MÉDIO REVENDA'].min().reset_index()

#RENOMEIA COLUNA
gr5.rename(columns={'PREÇO MÉDIO REVENDA': 'min'}, inplace=True)

#junta no elemento comum on=['DATA INICIAL', 'min']
preco_min = gr5.merge(max_diff, on=['DATA INICIAL', 'min'], how='inner')
preco_min.sort_values(by='diff_abs', ascending=False)

#seleciona os valores de interesse na tabela
preco_min = preco_min[['DATA INICIAL', 'ESTADO', 'min']].sort_values(by='DATA INICIAL', ascending=False)
```

```python
import plotly.express as px

fig = px.line(x=preco_min['DATA INICIAL'],
         y=preco_min['min'],
         labels={'x': 'Data', 'y': 'Preco'},
         width=1500,
         height=400,
         hover_name= preco_min['ESTADO'],
         markers=True,
         title='Menores valores registrados de gasolina comum')
 # Exibir o gráfico
fig.show()
print(preco_min)
```

```python
#agrupamento  para encontrar estados de valores maximos
gr7 =  gasolina_comum.groupby(['DATA INICIAL', 'ESTADO'])['PREÇO MÉDIO REVENDA'].max().reset_index()

#RENOMEIA COLUNA
gr7.rename(columns={'PREÇO MÉDIO REVENDA': 'max'}, inplace=True)

#junta no elemento comum on=['DATA INICIAL', 'min']
preco_max= gr7.merge(max_diff, on=['DATA INICIAL', 'max'], how='inner')
preco_max.sort_values(by='diff_abs', ascending=False)

#seleciona os valores de interesse na tabela
preco_max = preco_max[['DATA INICIAL', 'ESTADO', 'max']].sort_values(by='max', ascending=False)

```

```python
import plotly.express as px

fig = px.line(x=preco_max['DATA INICIAL'],
         y=preco_max['max'],
         labels={'x': 'Data', 'y': 'Preco'},
         width=1500,
         height=400,
         hover_name= preco_max['ESTADO'],
         markers=True,
         title='Maiores valores registrados de gasolina comum')
 # Exibir o gráfico
fig.show()
print(preco_max)
```
